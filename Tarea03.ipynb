### Ejercicio 1: Aritmética de corte (Chopping)

**Análisis:**
El método más preciso es **sumar en orden inverso** (desde el término más pequeño al más grande). En la aritmética de precisión finita (como el corte a 3 dígitos), al sumar un número muy pequeño a uno muy grande, los dígitos significativos del número pequeño se truncan y se pierden. Al sumar los números pequeños primero, estos se acumulan hasta ser lo suficientemente grandes como para afectar los dígitos significativos del total.

**Pseudocódigo:**
```text
Algoritmo Sumas_Corte
ENTRADA: n (número de términos)
SALIDA: suma_directa, suma_inversa

Paso 1: suma_directa = 0
Paso 2: PARA i DESDE 1 HASTA n HACER
            término = Cortar(1 / i^2, 3)
            suma_directa = Cortar(suma_directa + término, 3)

Paso 3: suma_inversa = 0
Paso 4: PARA i DESDE n HASTA 1 PASO -1 HACER
            término = Cortar(1 / i^2, 3)
            suma_inversa = Cortar(suma_inversa + término, 3)

Paso 5: RETORNAR suma_directa, suma_inversa

````python
import math

def truncar_3_digitos(val):
    if val == 0: return 0.0
    factor = 10 ** (3 - 1 - int(math.floor(math.log10(abs(val)))))
    return math.trunc(val * factor) / factor

print("--- Ejercicio 1 ---")
suma_dir = 0.0
for i in range(1, 11):
    term = truncar_3_digitos(1 / (i**2))
    suma_dir = truncar_3_digitos(suma_dir + term)
    
suma_inv = 0.0
for i in range(10, 0, -1):
    term = truncar_3_digitos(1 / (i**2))
    suma_inv = truncar_3_digitos(suma_inv + term)
    
print(f"Suma directa: {suma_dir}")
print(f"Suma inversa: {suma_inv}")
````

### Ejercicio 2: Serie de Maclaurin para arcotangente
La serie evaluada en $x=1$ es alternante: $4 \sum_{i=1}^{n} \frac{(-1)^{i+1}}{2i-1}$. En una serie alternante convergente, el error está acotado por el valor absoluto del primer término truncado. Por lo tanto, el error al usar $n$ términos es aproximadamente $\frac{4}{2n+1}$.

* **a.** Para un error $< 10^{-3}$:
  $\frac{4}{2n+1} < 10^{-3} \implies 4000 < 2n+1 \implies n > 1999.5$. Se necesitan **2000 términos**.
* **b.** Para un error $< 10^{-10}$:
  $\frac{4}{2n+1} < 10^{-10} \implies 4 \times 10^{10} < 2n+1 \implies n > 19999999999.5$. Se necesitan **$2 \times 10^{10}$ términos**.

### Ejercicio 5: Eficiencia de sumatorias
**a. Operaciones originales:**
La expresión $\sum_{i=1}^{n}\sum_{j=1}^{i} a_i b_j$ requiere sumar $1 + 2 + ... + n$ términos. Requiere **$\frac{n(n+1)}{2}$ multiplicaciones** y **$\frac{n(n+1)}{2} - 1$ sumas**.

**b. Modificación optimizada:**
Factorizando $a_i$ (ya que no depende de $j$): $\sum_{i=1}^{n} a_i \left( \sum_{j=1}^{i} b_j \right)$

**Pseudocódigo optimizado:**
```text
ENTRADA: n, arreglos a, b
SALIDA: suma_total

suma_total = 0
suma_b = 0
PARA i DESDE 1 HASTA n HACER:
    suma_b = suma_b + b[i]
    suma_total = suma_total + (a[i] * suma_b)
RETORNAR suma_total

# Algoritmo Raices_Cuadraticas
ENTRADA: a, b, c
SALIDA: x1, x2

* Paso 1: discriminante = b^2 - 4*a*c
* Paso 2: SI b > 0 ENTONCES
    * x1 = (-b - sqrt(discriminante)) / (2*a)
    * x2 = (2*c) / (-b - sqrt(discriminante))
    * SINO
        * x1 = (-b + sqrt(discriminante)) / (2*a)
        * x2 = (2*c) / (-b + sqrt(discriminante))
* Paso 3: RETORNAR x1, x2

# Código Python 



````python
print("\n--- Discusión 2: Raíces Cuadráticas ---")
def raices_cuadraticas(a, b, c):
    disc = b**2 - 4*a*c
    if disc < 0:
        return "Raíces complejas"
    if b > 0:
        x1 = (-b - math.sqrt(disc)) / (2*a)
        x2 = (2*c) / (-b - math.sqrt(disc))
    else:
        x1 = (-b + math.sqrt(disc)) / (2*a)
        x2 = (2*c) / (-b + math.sqrt(disc))
    return x1, x2

# Ejemplo con cancelación sustractiva latente
x1, x2 = raices_cuadraticas(1, 62.10, 1)
print(f"Raíz 1: {x1}")
print(f"Raíz 2: {x2}")

print("\n--- Discusión 3: Convergencia de la serie ---")
x = 0.25
tol = 1e-6
valor_real = (1 + 2*x) / (1 + x + x**2)
suma = 0.0
n = 1

while True:
    k1 = 2**(n-1)
    k2 = 2**n
    num = k1 * (x**(k1 - 1)) - k2 * (x**(k2 - 1))
    den = 1 - x**k1 + x**k2
    
    suma += num / den
    if abs(suma - valor_real) < tol:
        break
    n += 1

print(f"Valor real: {valor_real:.7f}")
print(f"Suma aprox: {suma:.7f}")
print(f"Se necesitaron {n} términos para alcanzar una diferencia menor a {tol}.")
````
